```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Criando a pipeline de pré-processamento dos dados

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# Selecionando as colunas numéricas
charging_num = charging.select_dtypes(include=['number']) # Define charging_num
# Criando o pipeline numérico
num_pipeline = Pipeline([
        ('std_scaler', StandardScaler()), # padronizando as escalas dos dados
    ])
# Aplicando a transformação
charging_num_tr = num_pipeline.fit_transform(charging_num)
```

```
charging_num_tr
```

```
try:
    from sklearn.compose import ColumnTransformer
except ImportError:
    from future_encoders import ColumnTransformer # Scikit-Learn < 0.20
```

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder # Importing OneHotEncoder

num_attribs = list(charging_num)
cat_attribs = ["region"]

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs), #tratando as variáveis numéricas (chamando a pipeline de cima)
        ("cat", OneHotEncoder(), cat_attribs), # tratando as variáveis categóricas
    ])

charging_prepared = full_pipeline.fit_transform(charging)
```

```
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(charging_prepared, charging_labels)
```

```
# vamos tentar o pipeline de pré-processamento completo em algumas instâncias de treinamento
some_data = charging.iloc[:5]
some_labels = charging_labels.iloc[:5]

some_data_prepared = full_pipeline.transform(some_data)

predictions = lin_reg.predict(charging_prepared)

print("Predictions:", lin_reg.predict(some_data_prepared))
```

```
print("Labels:", list(some_labels))
```

```
# avaliando o modelo

from sklearn.metrics import mean_squared_error
# erro médio quadrático eleva ao quadrado a média do erro médio absoluto. Estou avaliando se os erros não são tão grandes, esses erros são penalizados.
# penaliza muito mais valores distantes da média.

charging_predictions = lin_reg.predict(charging_prepared)
lin_mse = mean_squared_error(charging_labels, charging_predictions)
lin_rmse = np.sqrt(lin_mse) # raiz quadrada aqui
lin_rmse
```

Se a média dos valores é de aproximadamente 13k , esse modelo deu muito ruim , já que tem uma margem de erro de 11k

```
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(charging_labels, charging_predictions)
lin_mae
```

```
from sklearn.metrics import r2_score
```

```
r2 = r2_score(charging_labels, charging_predictions)
print('r²',r2)
```

```
def calculate_mape(labels, predictions):
    errors = np.abs(labels - predictions)
    relative_errors = errors / np.abs(labels)
    mape = np.mean(relative_errors) * 100
    return mape
```

```
# Calcular o MAPE
mape_result = calculate_mape(charging_labels, charging_predictions)

# Imprimir o resultado
print(f"O MAPE é: {mape_result:.2f}%")
```

```
from sklearn.tree import DecisionTreeRegressor

# Criando o modelo de DecisionTreeRegressor
model_dtr = DecisionTreeRegressor(max_depth=10)
model_dtr.fit(charging_prepared, charging_labels)
```
